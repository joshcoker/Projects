the  neurtral point of an aircraft, is teh point around which all aerodynamci forces are balenced. it is a good refrence point to compare to the cg. a conservative cg would be ahead of the np. 

neutral point = h0 + ns * vs *(as/aw)*(1-(de/da))
ho = aerodynimc center of wing typically 0.25
ns = stablizer efficenty, typcially 0.6
vs = stabilizer volume coefficent
as = lift curve slope of stabilizer
aw = lift curve slope of wing
de/da, change in stabilzer downwash angle versus changed in wing angle of attack, typically 0.t to 0.333

The tail-volume is calculated by multiplying the area of the stabiliser, including the elevator, by the distance between the tail and wing. This is called a volume because the results are in cubic inches or meters. The value is then divided by the area of the wing, and the average wing chord. Dividing by the wing area makes it relative to the wing area, a ratio to the wing area, and dividing by the wing chord makes the NP value relative to the chord, a percentage/fraction of the chord value.

vs = (s * l) / (sw *c)
where s is tail stabilier area 
l is tail arm lenght
sw is wing area
and c is average cord of wing

```python
import math

# wing chord
c = 4
# wing panel length (root to tip along sweep)
# the angled distance of lenght that is the wing for one side only
panel_length = 22
# sweep angle in degrees
sweep = 10

# stabilizer chord
sc = 3
# stabilizer span (assume straight, no sweep)
sg = 18
# read below markdown, on determing distance
tail_distance = 10.25
# --- Wing geometry ---
# projected half-span (horizontal component)
# half span is the wing span of just one wing. thats why wg = 2* half span
half_span = panel_length * math.cos(math.radians(sweep))
# full wingspan (projected)
wg = 2 * half_span
# wing area uses actual length (panel length × chord × 2)
wing_area = 2 * (panel_length * c)

# aspect ratio uses projected span
aspect = wg**2 / wing_area

# lift slope (2D to 3D) with sweep correction
anot = 0.111  # per degree lift coefficient (2D)
aw = (aspect * anot * math.cos(math.radians(sweep))) / (aspect + 18.25 * anot)

# --- Stabilizer geometry ---
saspect = sg / sc
snot = 0.095  # per degree lift coefficient stabilizer
ass = (saspect * snot) / (saspect + 18.25 * snot)

# volume coefficient
vsc = (sc * sg * tail_distance) / (wing_area * c)

# neutral point
np = (0.25 + (0.6 * vsc * (ass / aw) * (1 - 0.4)))
#cg in assembly minus distance from measure point to front of cord
dist = 3.846
rl = dist-4
gravcg = rl/c
statmarg = ((np - gravcg))*100
print("Neutral point (in chord fractions):", np)
print("tail stabilzer influnce,vsc",vsc)
print("tail stabilzer size percent of wing",(sc*sg)/wing_area)
print("center of gravty off of root wing",gravcg)
print("static margin",statmarg)
print("wing aspect ratio",aspect)

# stat margin is -48
```
